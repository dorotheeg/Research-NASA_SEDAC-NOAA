# T-Test Mexico Yearly

```python
from scipy import stats
import numpy as np
import rasterio
import gdal
from osgeo import gdal

import os
import glob
os.getcwd()
os.chdir(r'Z:\\')

from osgeo import gdal_array
from scipy.stats import chisquare
from scipy.stats import mannwhitneyu

import pymc3 as pm
import matplotlib.pyplot as plt
```

### 2012 vs 2018 for randomly chosen week (week 10)

```python
rasterArray_2012 = gdal_array.LoadFile(r'Z:\GEOVIIRS\data\mexico\agg_week\2012\week_2012_10.tif')
rasterArray_2012 = rasterArray_2012.flatten()


rasterArray_2018 = gdal_array.LoadFile(r'Z:\GEOVIIRS\data\mexico\agg_week\2018\week_2018_10.tif')
rasterArray_2018 = rasterArray_2018.flatten()


print(stats.ttest_ind(rasterArray_2012,rasterArray_2018, equal_var = False))

print("2012: ") 
print(chisquare(rasterArray_2012))
print("")
print("2018: ") 
print(chisquare(rasterArray_2018))
print("")
print("Compute the Kolmogorov-Smirnov statistic on 2 samples")
print(stats.ks_2samp(rasterArray_2012, rasterArray_2018))
print("")
print("Compute the Mann-Whitney rank test on samples x and y")
print(mannwhitneyu(rasterArray_2012, rasterArray_2018, use_continuity=True, alternative=None))
```
